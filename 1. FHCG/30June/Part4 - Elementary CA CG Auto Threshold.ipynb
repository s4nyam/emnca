```python
#@title Sort csv file

import csv
import re

def sort_csv_by_first_column(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        data = list(reader)
        headers = data[0]  # Get the headers
        data = data[1:]  # Skip the first row (headers)

        sorted_data = sorted(data, key=lambda row: extract_numeric_part(row[0]))

    new_csv_file = 'sorted_' + csv_file
    with open(new_csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)  # Write the headers back to the sorted file
        writer.writerows(sorted_data)

    print(f"Sorted data has been written to {new_csv_file}.")

def extract_numeric_part(text):
    numeric_part = re.findall(r'\d+', text)  # Extract numeric part using regex
    if numeric_part:
        return int(numeric_part[0])
    else:
        return 0

# Usage example
csv_filename = 'paths.csv'
sort_csv_by_first_column(csv_filename)
```

```python
#@title Display all results


import pandas as pd
from IPython.display import display, HTML
import base64
# Read the CSV file
df = pd.read_csv('sorted_paths.csv')

# Apply CSS style for zooming
display(HTML('''
<style>
.zoomable-image {
  transition: transform 0.2s;
}

.zoomable-image:hover {
  transform: scale(1.5);
}
</style>
'''))

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    html_string = ''
    html_string += '<div style="display: flex; flex-direction: row; margin: 10px;">'

    # Iterate over each image path in the row
    for col in ['path3', 'path2', 'path1']:
        image_path = row[col]
        image_data = base64.b64encode(open(image_path, 'rb').read()).decode('utf-8')

        html_string += '<div style="display: flex; flex-direction: column; text-align: center; margin-right: 10px;">'
        html_string += f'<p>Image: {image_path}</p>'

        # Add CSS class 'zoomable-image' to path2 image
        if col == 'path2':
            html_string += f'<img src="data:image/svg+xml;base64,{image_data}" class="zoomable-image" width="400" height="400" />'
        else:
            html_string += f'<img src="data:image/svg+xml;base64,{image_data}" width="400" height="400" />'

        html_string += '</div>'

    html_string += '</div>'

    display(HTML(html_string))
```
