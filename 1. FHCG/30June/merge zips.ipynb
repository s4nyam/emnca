```python
import os
import zipfile
import shutil

# Create a temporary directory to store the unzipped files
temp_dir = "/content/temp"
os.makedirs(temp_dir, exist_ok=True)

# Unzip all the zip files
for file in os.listdir("/content"):
    if file.endswith(".zip"):
        zip_path = os.path.join("/content", file)
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(temp_dir)

# Create a new zip file containing the unzipped files
combined_zip_path = "/content/combined_files.zip"
with zipfile.ZipFile(combined_zip_path, "w") as zip_ref:
    for folder_name, _, file_names in os.walk(temp_dir):
        for file_name in file_names:
            file_path = os.path.join(folder_name, file_name)
            zip_ref.write(file_path, os.path.relpath(file_path, temp_dir))

# Delete all files except the final combined zip file
for file in os.listdir("/content"):
    file_path = os.path.join("/content", file)
    if file_path != combined_zip_path and not os.path.isdir(file_path):
        os.remove(file_path)

# Remove the temporary directory
shutil.rmtree(temp_dir)
```
